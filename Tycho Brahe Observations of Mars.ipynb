### source files
The original file downloaded from the web site is `mars.xls`.

We translated this file into a tab separated file called `mars.txt` using Microsoft `Excel`



```python
!head -6 mars.txt
```

```python
df=pd.read_csv('mars.txt',sep='\t',skiprows=5)
df.head(3)
```

```python
df.columns
```

```python
core=df[['Days since 1 AD','Declination']]
core.head()
```

```python
core.plot('Days since 1 AD','Declination',style='.',figsize=[15,5])
ylabel('Declination')
grid()
```
